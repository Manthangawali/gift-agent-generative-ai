```python
gift_database = {
    "gaming": [
        ("RGB Mousepad", 300),
        ("Gaming Poster", 200),
        ("Retro Keychain", 150),
        ("Mini Joystick Toy", 250)
    ],
    "music": [
        ("Wireless Earbuds", 700),
        ("Bass Headphones", 900),
        ("Music Band Poster", 150),
        ("Portable Mini Speaker", 600)
    ],
    "books": [
        ("Motivational Book", 350),
        ("Notebook Set", 200),
        ("Bookmarks Set", 100)
    ],
    "fitness": [
        ("Hand Grip Strengthener", 250),
        ("Shaker Bottle", 200),
        ("Resistance Band", 350)
    ]
}
```

```python
def gift_scout_agent(recipient, interests, budget):
    budget = int(budget)

    suggestions = []
    
    for interest in interests:
        if interest in gift_database:
            for item, price in gift_database[interest]:
                if price <= budget:
                    suggestions.append((item, price, interest))

    if not suggestions:
        return "No suitable gift found under this budget."

    # Pick top 3
    suggestions = suggestions[:3]

    output = f"ðŸŽ Gift Ideas for {recipient} (Budget: â‚¹{budget})\n"
    output += "----------------------------------------\n"
    for item, price, interest in suggestions:
        output += f"â€¢ {item} â€” â‚¹{price} (matches interest: {interest})\n"

    return output
```

```python
print(gift_scout_agent(
    recipient="Brother, age 20",
    interests=["gaming", "music"],
    budget=1000
))
```
